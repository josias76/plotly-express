## Tache 2: DIAGRAMME A BARRE INTERACTIF

```python
# pays existant dans la base
gapminder["country"].unique()
```

```python
gapminder["country"].value_counts()
```

```python
# évolution de la population du canada , filtre en utilisant pandas
data_cana = gapminder [gapminder["country"]=="Canada"]
data_cana
```

```python
data_cana.shape
```

```python
# évolution de la population du canada , filtre en utilisant query de plotly
data_canada = px.data.gapminder().query("country == 'Canada'")
data_canada
```

```python
data_canada.shape
```

### évolution de la population du canada (image interactive)

```python
fig = px.bar(data_canada,x= 'year',y = 'pop', title = "Evolution de la population du Canada")
fig.show()
```

```python
# autre argument du px.bar

fig = px.bar(data_canada,x= 'year',y = 'pop',hover_data = ['lifeExp','gdpPercap'],color = 'lifeExp', title = "Evolution de la population du Canada")
fig.show()
```

## Tache 3 : NUAGE DE POINTS INTERACTIF (scatterplot)

```python
# Dataframe de l'année 2007
gapminder207= gapminder[gapminder['year']==2007]
gapminder207
```

```python
gapminder207.year.unique()
```

```python
gapminder2007 = gapminder.query("year == 2007")
gapminder2007
```

```python
# experance de vie en fonction du produit interieur brut
px.scatter(gapminder2007, x = "gdpPercap",y="lifeExp", title = "experance de vie en fonction du produit interieur brut pour 2007")
```

```python
gapminder.columns
```

```python
gapminder.continent.unique()
```

```python
# colorier les points par une variable catégorielle
px.scatter(gapminder2007, x = "gdpPercap",y="lifeExp",color = 'continent', title = "experance de vie en fonction du produit interieur brut pour 2007")
```

## Tache 4 : GRAPHIQUES INTERACTIFS A BULLES (BUBBLE CHART)

```python
# Répresentation d'une quatrième variable nuage de pointss precedent
px.scatter(gapminder2007, x = "gdpPercap",y="lifeExp",
           color = 'continent', 
           size = 'pop',
           title = "experance de vie en fonction du produit interieur brut pour 2007")
```

```python
# Ajout de la variable country
px.scatter(gapminder2007, x = "gdpPercap",y="lifeExp",
           color = 'continent', 
           size = 'pop',
           hover_name = 'country',
           title = "experance de vie en fonction du produit interieur brut pour 2007")
```

## Tache 5 :CREATION D'ANNIMATION INTERACTIVES

```python
# animation
px.scatter(gapminder, x = "gdpPercap",y="lifeExp",
           color = 'continent', 
           size = 'pop',
           hover_name = 'country',
           animation_frame = 'year',animation_group ='country',
           title = "experance de vie en fonction du produit interieur brut pour 2007")
```

```python
# Mettre axe x en  log
# on met les année en log

px.scatter(gapminder, x = "gdpPercap",y="lifeExp",
           log_x = True,
           color = 'continent', 
           size = 'pop',
           hover_name = 'country',
           animation_frame = 'year',animation_group ='country',
           title = "experance de vie en fonction du produit interieur brut pour 2007")
```

```python
# facet_col
# répresenter chaque continent dans son compartiment

px.scatter(gapminder, x = "gdpPercap",y="lifeExp",
           log_x = True,
           facet_col = 'continent',
           color = 'continent', 
           size = 'pop',
           hover_name = 'country',
           animation_frame = 'year' ,animation_group ='country',#var_catégorielle
           title = "experance de vie en fonction du produit interieur brut pour 2007")

#animation_frame = 'year' # var_temporaire,animation_group ='country',#var_catégorielle
```
